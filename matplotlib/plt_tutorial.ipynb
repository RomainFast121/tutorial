```python
import numpy as np
import matplotlib.pyplot as plt
```

# Scatter plot

```python
X_data = np.random.random(500) * 100 # values between 0 and 1 multiplied by 100
y_data = np.random.random(500) * 100

plt.scatter(X_data, y_data, c='steelblue', marker='*', s=100, alpha=0.5)
plt.show()
```

# Line plot

```python
years = [2006 + x for x in range(16)]
weights = [80 + 5 * np.round(x,1) for x in np.random.random(16)]

plt.plot(years, weights, c='b', linestyle='--', lw=1.5) 
plt.show()

plt.plot(years, weights, 'r-.', lw=1.5) 
plt.show()
```

# Bar plot

```python
x = ['C++', 'C#', 'Python', 'Java', 'Go']
y = [x for x in np.random.random(5)*20]

plt.bar(x, y, color='orange', align='center', width=0.8, edgecolor='purple', alpha=0.7, linewidth=2)
plt.show()
```

# Histogram

```python
ages = np.random.normal(40, 10, 500)

plt.hist(ages, bins=20)
plt.show()

plt.hist(ages, bins=[ages.min(), 18, 30, 40, 50, 60, ages.max()])
plt.show()

plt.hist(ages, bins=20, cumulative=True)
plt.show()
```

# Pie chart

```python
langs = ['C++', 'C#', 'Python', 'Java', 'Go']
votes = [50, 24, 14, 6, 17]
explode = [0, 0, 0, 0.3, 0]

plt.pie(votes, labels=langs, explode=explode, autopct='%.2f%%', pctdistance=0.7, startangle=90) # if not overlapping categories
plt.show()
```

# Boxplot

```python
heights = np.random.normal(172, 8, 500)

plt.boxplot(heights)
plt.show()

first = np.linspace(0,10,25)
second = np.linspace(10,200,25)
third = np.linspace(200,210,25)
fourth = np.linspace(210,230,25)

data = np.concatenate((first, second, third, fourth))
plt.boxplot(data)
plt.show()
```

# Plot customization

```python
years = [2006 + x for x in range(16)]
income = [50 + x for x in np.linspace(0,20,16) + np.random.normal(5,2,16)]
income_ticks = list(range(48,85,4))

plt.plot(years, income, c='darkred')
plt.title('Income of John (in USD)', fontsize=18, fontname='Academy Engraved LET', color='darkred')
plt.xlabel('Year', fontname='Academy Engraved LET', fontsize=12)
plt.ylabel('Yearly income in USD', fontname='Academy Engraved LET', fontsize=12)
plt.yticks(income_ticks, [f'${x}k' for x in income_ticks])

plt.show()
```

# Legends and multiple plots

```python
stock_a = [x for x in np.linspace(10, 50, 20) + np.random.normal(5,4,20)]
stock_b = [x for x in np.linspace(20, 70, 20) + np.random.normal(8,6,20)]
stock_c = [x for x in np.linspace(5, 60, 20) + np.random.normal(2,10,20)]

plt.plot(stock_a, label='stock_a')
plt.plot(stock_b, label='stock_b')
plt.plot(stock_c, label='stock_c')

plt.legend(loc='lower right')

plt.minorticks_on()
plt.grid(which='major', c='black', alpha=0.5)
plt.grid(which='minor', c='black', linestyle='--', alpha=0.3)

plt.show()
```

# Plot styling

```python
from matplotlib import style

style.use('bmh') # there are online list but can also create personal ones

langs = ['C++', 'C#', 'Python', 'Java', 'Go']
votes = [50, 24, 14, 6, 17]

plt.pie(votes, labels=None, startangle=90) 
plt.legend(labels=votes)
plt.show()
```

# Multiple figures

```python
style.use('default')

x1, y1 = np.random.random(100), np.random.random(100)
x2, y2 = np.arange(100), np.arange(100)

plt.figure(1)
plt.scatter(x1, y1)

plt.figure(2)
plt.scatter(x2, y2)

plt.show()
```

# Subplot

```python
x = np.arange(1,100)

fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title('Sin wave')

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title('Cos wave')

axs[1, 0].plot(x, np.random.random(len(x)))
axs[1, 0].set_title('Noise')

axs[1, 1].plot(x, np.log(x))
axs[1, 1].set_title('Log function')

plt.tight_layout() # avoid overlapping text 
fig.suptitle('Four plots')
plt.show()
```

# Exporting

```python
years = [2006 + x for x in range(16)]
income = [50 + x for x in np.linspace(0,20,16) + np.random.normal(5,2,16)]
income_ticks = list(range(48,85,4))

plt.plot(years, income, c='darkred')
plt.title('Income of John (in USD)', fontsize=18, fontname='Academy Engraved LET', color='darkred')
plt.xlabel('Year', fontname='Academy Engraved LET', fontsize=12)
plt.ylabel('Yearly income in USD', fontname='Academy Engraved LET', fontsize=12)
plt.yticks(income_ticks, [f'${x}k' for x in income_ticks])

plt.savefig('Export_test.png', dpi=200, bbox_inches='tight', transparent=True) 
# if dpi higher -> better quality, tight cut the fig, transparent removes the background

plt.close()
```

# 3D plotting

```python
ax = plt.axes(projection='3d')

x = np.arange(0, 50, 0.1)
y = np.sin(x)
z = np.cos(x)

ax.plot(x, y, z)
ax.set_title('3D Plot') # in a python file can rotate as well not in notebook
plt.show()

ax = plt.axes(projection='3d')

x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)

X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

ax.plot_surface(X, Y, Z, cmap='Spectral')
plt.show()
```
